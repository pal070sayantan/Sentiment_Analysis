```python
def missing_values_analysis(data):
  cols=[c for c in data.columns if data[c].isnull().sum()>0]
  miss_n=data[cols].isnull().sum().sort_values(ascending=True)
  ratio=(data[cols].isnull().sum()/data.shape[0]*100).sort_values(ascending=True)
  missing_val=pd.concat([miss_n,np.round(ratio,2)],axis=1,keys=["Missing_values","Ratio"])
  missing_val_df=pd.DataFrame(missing_val)
  return missing_val_df


def evaluate_dataframe(data,head=6,tail=6):
  print("SHAPE".center(82,'~'))
  print("Rows: {}".format(data.shape[0]))
  print("Column: {}".format(data.shape[1]))
  print("Types".center(82,'~'))
  print(data.dtypes)
  print("".center(82,'~'))
  print(missing_values_analysis(data))
  print("Duplicate Values".center(83,'~'))
  print(data.duplicated().sum())


evaluate_dataframe(data)



```

```python
def get_unique_counts_column(data):
  unique_val=pd.DataFrame({'Column_name': data.columns,'Count':(data[i].nunique() for i in data.columns)})
  unique_val=unique_val.sort_values('Count',ascending=False)
  unique_val=unique_val.reset_index(drop=True)
  return unique_val

get_unique_counts_column(data)
```

```python
constraints=['#B34D22','#EBE00C','#1FEB0C','#0C92EB','#EB0CD5']
def categorical_variable_analysis(data,column_name):
  fig=make_subplots(rows=1,cols=2,
                    subplot_titles=('CountPlot','Percentage'),
                    specs=[[{"type":"xy"},{'type':'domain'}]])#specs=[[{"types": "xy"},{'types':'domain'}]]


  fig.add_trace(go.Bar(y=data[column_name].value_counts().values.tolist(),
                       x=[str(i) for i in data[column_name].value_counts().index],
                       text=data[column_name].value_counts().values.tolist(),
                       textfont=dict(size=14),
                       name=column_name,
                       textposition='auto',
                       showlegend=False,
                       marker=dict(color=constraints,line=dict(color='#DBE6EC',width=1)
                                )),
                                row=1,col=1)

  fig.add_trace(go.Pie(labels=data[column_name].value_counts().keys(),
                       values=data[column_name].value_counts().values,
                       textfont=dict(size=20),
                       textposition='auto',
                       name=column_name,
                       showlegend=False,
                       marker=dict(colors=constraints)),
                row=1,col=2)

  fig.update_layout(title={'text': column_name,
                           'y':0.9,
                           'x':0.5,
                           'xanchor':'center',
                           'yanchor':'top'},
                    template='plotly_white')



  iplot(fig)
```

```python
categorical_variable_analysis(data,'overall')
```

```python
func=lambda x: re.sub("[^a-zA-Z]",' ',str(x))
data["reviewText"]=data["reviewText"].map(func)
data["reviewText"]=data["reviewText"].str.lower()
data.head()
```

```python
classifier = flair.models.TextClassifier.load('en-sentiment')
```

```python
from flair.data import Sentence
```

```python
for index,row in data.iterrows():
  text=row["reviewText"]
  sentence=Sentence(text)
  classifier.predict(sentence)
  sentiment=sentence.labels[0].value
  score=sentence.labels[0].score
  if(score<=0.8):
    sentiment='NEUTRAL'
  data.loc[index,'sentiment']=sentiment
  data.loc[index,'score']=score
```
